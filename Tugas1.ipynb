```python
import pandas as pd
wn = pd.read_csv("creditcard.csv")
```

```python
wn['Amount'].value_counts()
```

```python
from sklearn.preprocessing import  StandardScaler
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
x  = wn .apply(le.fit_transform)
x
```

```python
x1 = x[['V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10',
       'V11', 'V12', 'V13', 'V14', 'V15', 'V16', 'V17', 'V18', 'V19', 'V20',
       'V21', 'V22', 'V23', 'V24', 'V25', 'V26', 'V27', 'V28']]
```

```python
y1 = x[['Amount']].values
```

```python
x1
```

```python
y1[0:5]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split( x1, y1, test_size=0.1, random_state=0)
print ('Train set:', X_train.shape, Y_train.shape)
print ('Test set:', X_test.shape, Y_test.shape)
```

```python
import time
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,f1_score,recall_score
from sklearn.naive_bayes import GaussianNB
#from sklearn.svm import SVC

#from sklearn.neigh
#model = SVC(kernel='rbf',gamma='auto')
model = GaussianNB()
model.fit(X_train,Y_train)
y_val_pred = model.predict(X_test)
```

```python
print("accuracy score :",(accuracy_score (y_val_pred, Y_test))*100,"%")
```
